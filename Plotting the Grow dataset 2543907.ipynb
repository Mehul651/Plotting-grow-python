```python
#Importing necessary libraries 
import pandas as pd
import matplotlib.pyplot as plt 

#Step 1: Reading dataset and filtering 
df = pd.read_csv('GrowLocations.csv') 

#shows the dataset
df

#This line filters the data based on the Latitude and Longitude given in the question. 
#Latitude should range between 50.681 - 57.985 and Lonigtiude between -10.592 to 1.6848
df = df[(df['Latitude']. between(50.681, 57.985)) & (df['Longitude'].between(-10.592, 1.6848))] 

#shows the edited dataset 
df 

# Step 2: Overlay the Map with Coordinates
map_img = plt.imread('map7.png')   
# Display the map image without any axes
fig, ax = plt.subplots(figsize=(10, 10))  # Adjust the figsize as needed
ax.imshow(map_img)
ax.axis('off')  # Turn off the axes to display only the image

plt.show()

plt.scatter(df['Longitude'], df['Latitude'], c='red', marker='o', alpha=0.7, zorder=1) 

# This line Creates a plot of size 10x10
fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(map_img, extent=[-10.592, 1.6848, 50.681, 57.985])

for _, location in df.iterrows():
    ax.scatter(location['Longitude'], location['Latitude'], color='red', marker='o', s=5) 

# This line sets the tiltle and lables the x-axis and y-axis 
ax.set_title('Plotting grow data')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
#Display the plot 
plt.show()
```
